First I want a list of all events on ufcstats.com

```python
all_events_page = op.open_link('http://www.ufcstats.com/statistics/events/completed?page=all', async_=True)
all_events_soup = BeautifulSoup(all_events_page)

rows = all_events_soup.find_all(class_='b-statistics__table-content')[1:] #skip upcoming event
links = [row.find('a').get('href') for row in rows]
len(links)
```

```python
event_links = pd.DataFrame(links)
event_links
```

```python
event_links.to_csv('../data/ufcstats_events.csv')
```

Now I want all the bouts.

```python
event_link = event_links[0][0]
event_link
```

```python
event_page = op.open_link(event_link)
event_soup = BeautifulSoup(event_page)

bout_elems = event_soup.find_all(class_='b-fight-details__table-row b-fight-details__table-row__hover js-fight-details-click')

bout_links = [bout_elem.get('data-link') for bout_elem in bout_elems]

len(bout_links)
```

```python
bout_links
```

```python
bout_links = []

for event_link in event_links[0]:
    event_page = op.open_link(event_link)
    event_soup = BeautifulSoup(event_page)

    bout_elems = event_soup.find_all(class_='b-fight-details__table-row b-fight-details__table-row__hover js-fight-details-click')

    bout_links += [bout_elem.get('data-link') for bout_elem in bout_elems]

len(bout_links)
```

```python
pd.DataFrame(bout_links).to_csv('../data/ufcstats_bouts.csv')
```
