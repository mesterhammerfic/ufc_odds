# Loop through all links:

```python
for event_link in events.table.index[:1]:
    print('-------------------------')
    print(event_link)
    # Open Page
    event_page = op.open_tapology_link(event_link)
    event_soup = BeautifulSoup(event_page)
    
    # update bouts
    bouts.add_to_table(event_soup,event_link)
    # update events
    events.update_missing_info(event_soup,event_link)
    
    for bout_link in bouts.table.index:
        print('-------------------------')
        print('\t', bout_link)
        #open page
        bout_page = op.open_tapology_link(bout_link)
        bout_soup = BeautifulSoup(bout_page)
        
        # add to instances
        fighter_instances.add_to_table(bout_page, bout_link)
        # update bouts
        bouts.update_missing_info(bout_soup, bout_link)
        
        for fighter_link in fighter_instances.table.fighter_link:
            print('-------------------------')
            print('\t\t', fighter_link)
            #open page
            fighter_page = op.open_tapology_link(fighter_link)
            fighter_soup = BeautifulSoup(fighter_page)
            
            fighters.add_to_table(fighter_soup,fighter_link)
            
            
```
